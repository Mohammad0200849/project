```python
!pip install datasets --quiet

from datasets import load_dataset
import pandas as pd

dataset = load_dataset("dair-ai/emotion", split="train")
df = pd.DataFrame(dataset)
emotion_labels = dataset.features['label'].names
df['emotion'] = df['label'].apply(lambda x: emotion_labels[x])
surprise_texts = df[df['emotion'] == 'surprise']['text'].reset_index(drop=True)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Initialize tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(surprise_texts)

# Convert texts to sequences of integers
sequences = tokenizer.texts_to_sequences(surprise_texts)

# Get max sequence length
max_len = max(len(seq) for seq in sequences)

# Pad sequences
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

# Vocabulary size for GAN generator/discriminator
vocab_size = len(tokenizer.word_index) + 1
```

```python
# Convert token IDs to words using tokenizer
for i, seq in enumerate(generated_ids):
    words = [tokenizer.index_word.get(token, '') for token in seq]
    sentence = ' '.join([word for word in words if word.strip() != ''])
    print(f"Generated Sample {i+1}:\n{sentence}\n" + "-"*60)
```

```python
synthetic_texts = []  # store generated text samples here

for seq in generated_ids:
    words = [tokenizer.index_word.get(token, '') for token in seq]
    sentence = ' '.join([word for word in words if word.strip() != ''])
    synthetic_texts.append(sentence)
```

```python
# Original imbalanced dataset
df_balanced = df[df['emotion'] != 'surprise']  # all other emotions
surprise_real = df[df['emotion'] == 'surprise']

# Convert synthetic surprise texts to DataFrame
synthetic_df = pd.DataFrame({'text': synthetic_texts, 'emotion': 'surprise'})

# Combine real data + synthetic surprise to form balanced dataset
augmented_df = pd.concat([df_balanced, surprise_real, synthetic_df], ignore_index=True)
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Tokenize and pad
X = tokenizer.texts_to_sequences(augmented_df['text'])
X = pad_sequences(X, maxlen=sequence_length)

# Encode labels
label_map = {label: i for i, label in enumerate(augmented_df['emotion'].unique())}
y = augmented_df['emotion'].map(label_map)
y = to_categorical(y)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GlobalAveragePooling1D, Dense

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=sequence_length))
model.add(GlobalAveragePooling1D())
model.add(Dense(64, activation='relu'))
model.add(Dense(y.shape[1], activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

# Predict
y_pred = model.predict(X_test)
y_pred_labels = np.argmax(y_pred, axis=1)
y_true_labels = np.argmax(y_test, axis=1)
```

```python
print("Classification Report:\n")
print(classification_report(y_true_labels, y_pred_labels, target_names=label_map.keys()))
```
